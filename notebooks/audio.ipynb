# Procedure

```python
# import data
wav_filename = "/Users/emilydennis/Desktop/210522_011.WAV"
sample_rate, wav_file = wavfile.read(wav_filename)
```

```python
subset1 = wav_file[:5*sample_rate,0]
```

```python
plt.plot(subset1)
```

```python
frequencies, times, spectrogram = ss.spectrogram(subset1, sample_rate)
```

```python
plt.pcolormesh(times,frequencies,spectrogram,shading='nearest')
plt.ylim([0,10000])
```

```python
test = np.fft.fft(spectrogram)
```
